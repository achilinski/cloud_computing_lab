```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

df = pd.read_csv("summary.csv")

# Convert response times from ms to seconds
df["min_response_time_sec"] = df["min_response_time"] / 1000
df["average_response_time_sec"] = df["average_response_time"] / 1000
df["max_response_time_sec"] = df["max_response_time"] / 1000

# Plot response times vs client count
plt.figure(figsize=(10, 6))

plt.plot(df["clients"], df["min_response_time_sec"], marker="o", label="Min Response Time (s)")
plt.plot(df["clients"], df["average_response_time_sec"], marker="o", label="Average Response Time (s)")
plt.plot(df["clients"], df["max_response_time_sec"], marker="o", label="Max Response Time (s)")

plt.xlabel("Number of Clients")
plt.ylabel("Response Time (seconds)")
plt.title("Response Time vs Client Count")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("summary.csv")

# Sort by client count (recommended)
df = df.sort_values("clients")

# Plot request counts vs client count
plt.figure(figsize=(10, 6))

plt.plot(
    df["clients"],
    df["overall_request_count"],
    marker="o",
    label="Total Requests"
)

plt.plot(
    df["clients"],
    df["overall_failure_count"],
    marker="o",
    label="Failed Requests"
)

plt.xlabel("Number of Clients")
plt.ylabel("Number of Requests")
plt.title("Request Count vs Client Count")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
